#### **Dot Product using matrix 1x1**

$$
\mathbf{a} \cdot \mathbf{b} = 
\begin{bmatrix}
a_1 & a_2 & \cdots & a_n
\end{bmatrix}
\begin{bmatrix}
b_1 \\
b_2 \\
\vdots \\
b_n
\end{bmatrix}
= a_1b_1 + a_2b_2 + \cdots + a_nb_n
$$


#### **Dot Product using matrix 2x2**
$$
\mathbf{A} \cdot \mathbf{B} = 
\begin{bmatrix}
a_{11} & a_{12} \\
a_{21} & a_{22}
\end{bmatrix}
\begin{bmatrix}
b_{11} & b_{12} \\
b_{21} & b_{22}
\end{bmatrix}
=
\begin{bmatrix}
a_{11}b_{11} + a_{12}b_{21} & a_{11}b_{12} + a_{12}b_{22} \\
a_{21}b_{11} + a_{22}b_{21} & a_{21}b_{12} + a_{22}b_{22}
\end{bmatrix}
$$


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#### Example of Matrix dot product in numpy
def activation(w: np.ndarray, b: np.ndarray, x: np.ndarray) -> np.ndarray:
    b = b.reshape(-1, 1)
    w_shape, b_shape, x_shape = w.shape, b.shape, x.shape

    if w_shape[1] == x_shape[0]:
        if b_shape[0] == w_shape[0]:
            product = b + np.dot(w, x)
            return product

if __name__ == "__main__":
    w = np.array([[1, 4], [2, -1], [0.5, 3]])
    b = np.array([[9], [6], [3]])
    x = np.array([[7, 1, 0, 2], [9, -1, 0.5, 1]])
    A = activation(w, b, x)
    print(f"A matrix shape: {A.shape}")
    print(A)
```

#### **Single Layer Perceptrons**


$$
\mathbf{x} = 
\begin{bmatrix}
x_1 \\
x_2 \\
\vdots \\
x_n
\end{bmatrix}, \quad
\text{and weight vector:} \quad
\mathbf{w} =
\begin{bmatrix}
w_1 \\
w_2 \\
\vdots \\
w_n
\end{bmatrix}
$$

$$
z = \mathbf{w}^T \mathbf{x} + b
= w_1 x_1 + w_2 x_2 + \cdots + w_n x_n + b
$$

$$
y = \sigma(z) = \sigma(\mathbf{w}^T \mathbf{x} + b)
$$

#### **Multi Layer Perceptrons (MLP)**

$$
b + W^T \cdot x
$$

$$
W^T = \begin{bmatrix} W_1 & W_2 & W_3 \end{bmatrix}
\quad
X = \begin{bmatrix} x_1 \\ x_2 \\ x_3 \end{bmatrix}
$$

$$
a_1 = W_{11}x_1 + W_{12}x_2 + b_1
$$
$$
a_2 = W_{21}x_1 + W_{22}x_2 + b_2
$$

$$
z_1 = h(a_1^{(1)})
$$
$$
z_2 = h(a_2^{(1)})
$$

$$
y = h(a^{(2)}) = V_1z_1 + V_2z_2 + b
$$

$$
X = \begin{bmatrix} x_{11} & x_{12} \\ x_{21} & x_{22} \\ x_{31} & x_{32} \\ x_{41} & x_{42} \end{bmatrix} = \begin{bmatrix} 0 & 0 \\ 1 & 0 \\ 0 & 1 \\ 1 & 1 \end{bmatrix}
$$

$$
W = \begin{bmatrix} W_{11} & W_{12} \\ W_{21} & W_{22} \end{bmatrix} = \begin{bmatrix} 1 & 1 \\ 2 & 2 \end{bmatrix}
$$

$$
b = \begin{bmatrix} b_1 \\ b_2 \end{bmatrix} = \begin{bmatrix} -2 \\ -2 \end{bmatrix}

$$
A = W \cdot X^T = \begin{bmatrix} W_{11} & W_{12} \\ W_{21} & W_{22} \end{bmatrix} \begin{bmatrix} x_{11} & x_{21} & x_{31} & x_{41} \\ x_{12} & x_{22} & x_{32} & x_{42} \end{bmatrix}
$$

$$
A = \begin{bmatrix} W_{11}x_{11} + W_{12}x_{12} & W_{11}x_{21} + W_{12}x_{22} & W_{11}x_{31} + W_{12}x_{32} & W_{11}x_{41} + W_{12}x_{42} \\ W_{21}x_{11} + W_{22}x_{12} & W_{21}x_{21} + W_{22}x_{22} & W_{21}x_{31} + W_{22}x_{32} & W_{21}x_{41} + W_{22}x_{42} \end{bmatrix}
$$

$$
A = \begin{bmatrix} t_{11} & t_{12} & t_{13} & t_{14} \\ t_{21} & t_{22} & t_{23} & t_{24} \end{bmatrix} + \begin{bmatrix} b_1 & b_1 & b_1 & b_1 \\ b_2 & b_2 & b_2 & b_2 \end{bmatrix}
$$

$$
A = W \cdot X^T + b \cdot 1^T
$$

**Note the dimensions: (M x N) = (M x D) * (D x N) + (M x 1)**
$$
A = W \cdot X^T + b
$$


$$
Z = \begin{bmatrix} h(a_{11}) & h(a_{12}) & \dots \\ h(a_{21}) & h(a_{22}) & \dots \end{bmatrix} = h(A)
$$

**Note the dimensions: (K x N) = (K x M) * (M x N) + (K x 1)**
$$
d = V^T Z + C
$$

$$
Y = \begin{bmatrix} h(d_{11}) & h(d_{12}) & \dots \\ h(d_{21}) & h(d_{22}) & \dots \end{bmatrix}
$$

**Trapose matrix X^T (D x N)**
$$
X^T = \begin{bmatrix} 0 & 1 & 0 & 1 \\ 0 & 0 & 1 & 1 \end{bmatrix}
$$

**Weight**
$$
V = \begin{bmatrix} 3 & -3 \end{bmatrix} \quad C = \begin{bmatrix} -3 \end{bmatrix}
$$

**Output**
$$
Y = \begin{bmatrix} 0 & 1 & 1 & 0 \end{bmatrix}
$$

#### **MLP Example**

```python
def sigmoid(a):
    return 1/(1 + np.exp(-a))

def hardLimit(a):
    return np.where(a >= 0, 1, 0)
```

```python
plt.figure(figsize=(6,4))
xs = np.linspace(-10, 10, 500)
plt.title("Sigmoid Function")
plt.plot(xs, sigmoid(xs))
plt.show()
```

```python
def mlp(X, W, B, h1, V, C, h2):
    A = W @ X + B
    z = h1(A)
    D = V @ z + C
    Y = h2(D)

    return Y
def mlp2(X, W, B, V, C, h=sigmoid, g=lambda a: a):
    A = B + np.dot(W, X)
    Z = h(A)
    D = C + np.dot(V, Z)
    Yp = g(D)
    return Yp
```

```python
x = np.array([[0,1,0,1], 
              [0,0,1,1]])

w = np.array([[1,1],
              [1,2]])

b = np.array([[-2], [-2]])
v = np.array([[-3, 3]])
c = np.array([[-3]])

print(x.shape)
print(w.shape)
print(b.shape)
print(v.shape)
print(c.shape)
```

```python
#### Using hardlimit function
y = mlp(x, w, b, hardLimit, v, c, hardLimit)
y
```

```python
##### Using sigmoid function
y = mlp2(x, w, b, v, c, h=sigmoid, g=sigmoid)
y
```

#### **Single Layer Perceptron Example**

```python
class Perceptron:
    def __init__(self, inputs, lr = 0.1):
        self.weight = np.zeros(inputs)
        self.bias = 0
        self.lr = lr

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))
    
    def sigmoid_derivative(self, x):
        return x * (1 - x)

    def foward_prop(self, x):
        res = np.dot(x, self.weight) + self.bias
        return self.sigmoid(res)
    
    def train(self, x, y, epochs=500): 
        for i in range(epochs):
            out = self.foward_prop(x)
            err = y - out

            d_out = err * self.sigmoid_derivative(out)
            self.weight += self.lr * np.dot(x.T, d_out)
            self.bias += self.lr * np.sum(d_out)

if __name__ == "__main__":
    x = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
    y = np.array([0, 0, 0, 1])

    p = Perceptron(inputs=2, lr = 0.1)
    p.train(x, y, epochs=1000)

    for i in x:
        pred = p.foward_prop(i)
        print(f"Input: {i} -> Predicted: {pred:.4f}")
    
    print(f"\nWeight: {p.weight}")
    print(f"Final bias: {p.bias}")
    
```

### **Back Propagation**


At first we started with...

\begin{align}
\delta_{kn}^{(2)} &= \frac{\partial E_n}{\partial d_{kn}}
\\
&= \frac{1}{2}\frac{\partial \sum_r (\hat{y}_{rn} - y_{rn})^2}{\partial d_{kn}}
= \frac{1}{2} \sum_r \frac{\partial (\hat{y}_{rn} - y_{rn})^2}{\partial d_{kn}}
= \sum_r (\hat{y}_{rn} - y_{rn}) \frac{\partial (\hat{y}_{rn} - y_{rn})}{\partial d_{kn}}
\\
&= 0 + \cdots + 0 + (\hat{y}_{kn} - y_{kn}) \frac{\partial \hat{y}_{kn}}{\partial d_{kn}} + 0 + \cdots + 0
\\
&= \hat{y}_{kn} - y_{kn}
\end{align}

Since $a_{jn}$ affects $E_n$ through layer-2 nodes, then with the chain rule

\begin{align}
\frac{\partial E_n}{\partial a_{jn}}
&= \sum_k \frac{\partial E_n}{\partial d_{kn}} \cdot \frac{\partial d_{kn}}{\partial a_{jn}}
= = \sum_k \delta_{kn}^{(2)} \cdot \frac{\partial d_{kn}}{\partial a_{jn}}
\\
&= \sum_k \delta_{kn}^{(2)} \cdot \frac{\partial (c_k + \sum_r v_{kr} z_{rn})}{\partial a_{jn}} = \sum_k \delta_{kn}^{(2)} \cdot v_{kj}
\frac{\partial z_{jn}}{\partial a_{jn}}
\\
&= \sum_k \delta_{kn}^{(2)} \cdot v_{kj}
\frac{\partial h(a_{jn})}{\partial a_{jn}}
= \sum_k \delta_{kn}^{(2)} \cdot v_{kj} h'(a_{jn})
\\
&= h'(a_{jn}) \sum_k \delta_{kn}^{(2)} \cdot v_{kj}
\end{align}


Let consider $h'(a) = \frac{d h(a)}{d a}$.

For sigmoid,

\begin{align}
h'(a) &= \frac{d h(a)}{d a} = \frac{d \frac{1}{1 + \exp(-a)}}{d a}
\\
&= \frac{\exp(-a)}{(1 + \exp(-a))^2}
\end{align}

Re-arrange to make it nicer.
\begin{align}
h'(a) &=
\left( \frac{1}{1 + \exp(-a)} \right) \left(\frac{\exp(-a)}{1 + \exp(-a)}\right)
\\
&= h(a) \cdot \left(\frac{1 + \exp(-a) - 1}{1 + \exp(-a)}\right)
= h(a) \cdot \left(\frac{1 + \exp(-a)}{1 + \exp(-a)} - \frac{1}{1 + \exp(-a)}\right)
\\
&= h(a) \cdot \left(1 - h(a) \right)
\end{align}


#### Layer-2 bias.

\begin{align}
\frac{\partial E_n}{\partial c_k} &= \frac{\partial E_n}{\partial d_{kn}} \cdot \frac{\partial d_{kn}}{\partial c_k}
\\
&= \delta_{kn}^{(2)} \frac{\partial (c_k + \sum_j v_{kj} z_{jn})}{\partial c_k}
\\
&= \delta_{kn}^{(2)}
\end{align}
for $k = 1, \ldots, K$.


#### Layer-2 weight.

\begin{align}
\frac{\partial E_n}{\partial v_{kj}}
&= \frac{\partial E_n}{\partial d_{kn}} \cdot \frac{\partial d_{kn}}{\partial v_{kj}}
\\
&= \delta_{kn}^{(2)} \frac{\partial (c_k + \sum_r v_{kr} z_{rn})}{\partial v_{kj}}
\\
&= \delta_{kn}^{(2)} \cdot z_{jn}
\end{align}

#### Layer-1 bias.

\begin{align}
\frac{\partial E_n}{\partial b_j} &= \frac{\partial E_n}{\partial a_{jn}} \cdot \frac{\partial a_{jn}}{\partial b_j}
\\
&= \delta_{jn}^{(1)} \frac{\partial (b_j + \sum_i w_{ji} x_{in})}{\partial b_j}
\\
&= \delta_{jn}^{(1)}
\end{align}
for $j = 1, \ldots, M$.

#### Layer-1 Weight

\begin{align}
\frac{\partial E_n}{\partial w_{ji}} &= \frac{\partial E_n}{\partial a_{jn}} \cdot \frac{\partial a_{jn}}{\partial w_{ji}}
\\
&= \delta_{jn}^{(1)} \frac{\partial (b_j + \sum_r w_{jr} x_{rn})}{\partial w_{ji}}
\\
&= \delta_{jn}^{(1)} \cdot x_{in}
\end{align}
for $j = 1, \ldots, M$.

#### **ANN with sklearn library**

```python
from sklearn.linear_model import Perceptron
from sklearn.neural_network import MLPClassifier, MLPRegressor
from sklearn.metrics import accuracy_score, mean_squared_error
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
iris = load_iris()
x = iris.data
y = iris.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

mlp = MLPClassifier(
    hidden_layer_sizes=(10, 5),
    max_iter=1000,
    activation="relu",
    solver="adam",
    random_state=0
)

mlp.fit(x_train, y_train)
```

```python
y_pred = mlp.predict(x_test)
acc = accuracy_score(y_test, y_pred)
print(f"Accuracy: {acc:.4f}")
```

```python
pred_flower = [iris.target_names[p] for p in y_pred[:5]]
actual = [iris.target_names[a] for a in y_test[:5]]

flower_pred_data = {
    "Predicted": pred_flower,
    "Actual": actual
}

df = pd.DataFrame(flower_pred_data)
display(df)
```

```python
##### Create the onehot encoder

def onehot(y, classes=None):
  if classes is None:
    classes = np.unique(y)

  K = len(classes)
  N = y.shape[0]
  codebook = np.diag(np.ones(K))

  yhot = codebook[y[:,0]]

  return yhot
```
