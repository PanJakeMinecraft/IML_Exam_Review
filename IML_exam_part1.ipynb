### **Exercise 1:**
Let define gray as having
\begin{align}
|B - G| \leq \tau \\
|B - R| \leq \tau \\
|G - R| \leq \tau \\
\end{align}

Let set $\tau$ as 80.

````python
#### Solution exercise 1
im2 = image.copy()
tau = 80

mask = (np.abs(im2[:,:,0] - im2[:,:,1]) <= tau) & \
       (np.abs(im2[:,:,0] - im2[:,:,2]) <= tau) & \
       (np.abs(im2[:,:,1] - im2[:,:,2]) <= tau)

im2_masked = np.zeros_like(im2)
im2_masked[mask] = im2[mask]  

plt.imshow(im2_masked)
plt.axis("off")
plt.show()
````

## **PART2: CSE and Scipy**

**Given the equation**
* $A + B + C = 3000000$
* $A - C = 400000$
* $B = 1.2 C$

Writing in matrix equation
$$
\begin{align}
\begin{bmatrix}
1 & 1 & 1\\
1 & 0 & -1 \\
0 & 1 & -1.2 \\
\end{bmatrix}
\cdot
\begin{bmatrix}
A \\ B \\ C
\end{bmatrix}
=
\begin{bmatrix}
3000000 \\ 400000 \\ 0
\end{bmatrix}
\end{align}
$$

In this approach, solve the equation and find the value of x that satisfy all equations in the system

$\mathbf{A} \cdot \mathbf{x} = \mathbf{b}$.

````python
a = np.array([[1, 1, 1],
              [1, 0, -1],
              [0, 1, -1.2]])

b = np.array([[3000000],
              [400000],
              [0]])

res = np.linalg.solve(a, b)
res
````

### **Finding Eigenvectors and eigenvalues**

**Find the eigenvalues**

$$
A = \begin{bmatrix} 4 & 2 \\ 1 & 3 \end{bmatrix}
$$


$$
\det(A - \lambda I) = 0
$$

$$
A - \lambda I = 
\begin{bmatrix} 4-\lambda & 2 \\ 1 & 3-\lambda \end{bmatrix}
$$


$$
\det(A - \lambda I) = (4-\lambda)(3-\lambda) - (2)(1) = 0
$$

$$
(4-\lambda)(3-\lambda) - 2 = 0
$$


$$
12 - 4\lambda - 3\lambda + \lambda^2 - 2 = 0
$$

$$
\lambda^2 - 7\lambda + 10 = 0
$$


$$
\lambda = \frac{7 \pm \sqrt{(-7)^2 - 4 \cdot 1 \cdot 10}}{2} = \frac{7 \pm \sqrt{49 - 40}}{2} = \frac{7 \pm 3}{2}
$$

$$
\lambda_1 = 5, \quad \lambda_2 = 2
$$

**Find the eigenvector**

$$
(A - \lambda I)v = 0
$$

$$
A - 5I = \begin{bmatrix} -1 & 2 \\ 1 & -2 \end{bmatrix}
$$

$$
\begin{bmatrix} -1 & 2 \\ 1 & -2 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix}
$$


$$
-x + 2y = 0 \implies x = 2y
$$

Let $y=1 \implies x = 2$  

$$
v_1 = \begin{bmatrix} 2 \\ 1 \end{bmatrix}
$$


$$
A - 2I = \begin{bmatrix} 2 & 2 \\ 1 & 1 \end{bmatrix}

$$
\begin{bmatrix} 2 & 2 \\ 1 & 1 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix}

$$

$$

Let $y=1 \implies x = -1$  

$$

### **Scipy**

````python
import scipy as sp
````

````python
points = np.array([[100, 0], [100, 100], [0, 100], [0, 0], [-20, 50]])
ground_truth = np.array((30, 70)).reshape((1,2))
D = np.sqrt(np.sum((points - ground_truth)**2, axis = 1))
D
````

````python
strength = 1/D ** 2
strength
````

````python
def loss(P, Rs, Ds):
    '''
    P : unknown position, np.array (1, 2)
    Rs: reference positions, np.array (N, 2)
    Ds: distances from the uknown position to references, np.array (N, 1)
    '''
    P = np.array(P).reshape((1,2))
    Ds = np.array(Ds).reshape((-1,1))
    c = np.sum((Rs - P)**2, axis=1).reshape((-1,1)) - Ds**2
    loss = np.sum(c**2)
    return loss

a = loss(ground_truth, points, D.reshape((-1,1)))
a ##### this is for testing the code
````

````python
##### Now, we will minimize
def minimize_loss(x):
    return loss(x.reshape((1,2)), points, D.reshape(-1,1))

P0 = np.array([1,0]) #### prediction points
res = sp.optimize.minimize(minimize_loss, x0=P0)
res
````

````python
#### Derivaties
def f(x):
    return x**3 + x**2 + 3*x + 3

def g(x):
    return 2 * np.sin(x) - ((x)**2 / 10)

diff = sp.misc.derivative(f, 1, n=1) #### 1st derivative
diff2 = sp.misc.derivative(g, 2.5, n=2) # 2nd order derivative
print(diff)
print(diff2)
````

````python
##### integrations
def f(y, x):
    return np.sin(x + y**2)

def y_lower(x):
    return -x

def y_upper(x):
    return x ** 2

res, err = sp.integrate.dblquad(f, 0, 1, y_lower, y_upper)
print(f"Integration result: {res}")
print(f"Integration error: {err}")
````

### **Exercise 3:**
Make 2x3 subplots
Each plot with ```'equal'``` aspect ratio.

* The 1st plot shows points $\mathrm{P}^T$ ($4 \times 2$; x-coordinates in the 1st column; y-coorindates in the 2nd column),

| x | y |
|---|---|
| 1 | 1 |
| 1 | -1 |
| -1 | -1 |
| -1 | 1 |

* The 2nd-6th plots show points $( \mathrm{A} \cdot \mathrm{P}^T )^T$
where
$\mathrm{A} = \begin{bmatrix} \cos(\theta) & -\sin(\theta) \\ \sin(\theta) & \cos(\theta) \end{bmatrix}$ for $\theta = \pi/10, \pi/8, \pi/6, \pi/4, \pi/2$, respectively.

````python
P = np.array([[1, 1],
              [1, -1],
              [-1, -1],
              [-1, 1]])

def matA(th):
    return np.array([[np.cos(th), -np.sin(th)],
                     [np.sin(th),  np.cos(th)]])

thetas = [np.pi/10, np.pi/8, np.pi/6, np.pi/4, np.pi/2]

fig, axes = plt.subplots(2, 3, figsize=(8,4))
fig.tight_layout(h_pad=4, w_pad=4)

##### the first graph
axes[0,0].plot(P[:,0], P[:,1], "ro-")
axes[0,0].axis("equal")
axes[0,0].set_xlim(-3,3)
axes[0,0].set_ylim(-3,3)
axes[0,0].set_title("Original \Graph")

#### the 2nd to 6th graph
for i, th in enumerate(thetas):
    A = matA(th)
    APT = A @ P.T
    
    plt.subplot(2, 3, 2+i)
    plt.plot(APT.T[:, 0], APT.T[:, 1], "go-")
    plt.plot(0, 0, "k*")
    plt.axis("equal")
    plt.xlim(-3,3)
    plt.ylim(-3,3)
    plt.title("Graph {}".format(2+i))

plt.show()
````
